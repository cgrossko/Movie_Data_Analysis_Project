```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv('/Users/christiangrosskopf/Downloads/movies.csv')
```

```python
df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}  -  {}%'.format(col, pct_missing))
```

```python
# Exclude movies with incomplete data from analysis by removing those with a value of 0 or Nan

df = df[(df != 0).all(axis=1)]
```

```python
# Data types for columns

df.dtypes
```

```python
# Change budget and gross columns to accounting datatype
df.fillna(0, inplace=True)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# sort data by gross

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Correlation by budget, company

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs. Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget of Films')
plt.show()
```

```python
# Plot budget vs. gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
numeric_df = df.select_dtypes(include=['float', 'int'])

correlation_matrix = numeric_df.corr(method='pearson')
print(correlation_matrix)

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# correlation by company

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
